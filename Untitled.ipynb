```python
import datetime
from dateutil.relativedelta import relativedelta
from opt_assistant import Opt_Assistant
opt_info = Opt_Assistant()
from stock_centre import *
import matplotlib.pyplot as plt
hq = Stock_Data_Centre()

ts_obj = TuShare()
pro = ts_obj.pro
```

```python
# 回测时间不要超过两年

start_date = '2021-07-01'
# str: 回测开始时间‘%Y-%m-%d’格式

end_date = '2023-06-30'
# str: 回测结束时间‘%Y-%m-%d’格式，不超过'2023-06-30'

init_cash = 1000000
# int, 账户初始资金

underlying_symbol = '510500'
# str: 标的代码，只输入数字无需包含交易所代码，现允许所有金融期权三类指数九类ETF，手续费根据类型自动调整
# 指数类：沪深300 '000300'；中证1000 '000852'；上证50 '000016'
# ETF类：上证50ETF '510050'；沪深300ETF '510300'；中证500ETF '510500'；
# 科创50ETF '588000'；科创板50ETF '588080'；
# 创业板ETF '159915'；深证100ETF '159901'；沪深300ETF '159919'；中证500ETF '159922'

time_gear = 1
# int, 期权到期月份，如1为下月到期，现不支持开当月合约，范围1,2,3,6,9,12

date_select = 5
# int, 调仓日为每月几号，现仅支持开平同一天

# 以下参数支持多腿组合构建，列表第一个数为第一腿参数，以此类推

trade_num = [-1,1,2] 
# list, 交易手数

price_gear = [1,-1,1]
# list, 档位，负数为行权价小于标的价，正数为行权价大于标的价档位，每个元素范围-3,-2,-1,1,2,3

contract_type = ['CO','CO','PO']
# list, 期权类型，每个元素范围’CO’, ‘PO’

# 组合案例（均为买入情况）
# 牛市看涨：trade_num = [正，负] price_gear = [小，大] contract_type = ['CO', 'CO']
# 牛市看跌：trade_num = [正，负] price_gear = [小，大] contract_type = ['PO', 'PO']
# 熊市看涨：trade_num = [正，负] price_gear = [大，小] contract_type = ['CO', 'CO']
# 熊市看跌：trade_num = [正，负] price_gear = [大，小] contract_type = ['PO', 'PO']
# 跨式/宽胯：trade_num = [正，正] price_gear = [大，小，或相同] contract_type = ['CO', 'PO']
# 领式：trade_num = [正，负] price_gear = [小，大] contract_type = ['PO', 'CO']
```

```python
#  标的范围
index_list = ['000300', '000852', '000016']
etf_list = ['510050', '510300', '510500', '588000', '588080', '159915', '159901', '159919', '159922']
#  标的历史价格
if underlying_symbol in index_list:
    underlying_symbol_for_stock = underlying_symbol+'.SH'
    underlying_symbol_for_opt = underlying_symbol+'.XSHG'
    underlying_prc_his = hq.get_hq(code=underlying_symbol_for_stock, start_date=start_date, end_date=end_date, index_data=True)
elif underlying_symbol in etf_list:
    if underlying_symbol.startswith('5'):
        underlying_symbol_for_stock = underlying_symbol+'.SH'
        underlying_symbol_for_opt = underlying_symbol+'.XSHG'        
    elif underlying_symbol.startswith('1'):
        underlying_symbol_for_stock = underlying_symbol+'.SZ'
        underlying_symbol_for_opt = underlying_symbol+'.XSHE'
    underlying_prc_his_Tushare = pro.fund_daily(ts_code=underlying_symbol_for_stock, start_date=pd.to_datetime(start_date, format='%Y-%m-%d').strftime('%Y%m%d'), end_date=pd.to_datetime(end_date, format='%Y-%m-%d').strftime('%Y%m%d'))
    underlying_prc_his = underlying_prc_his_Tushare.set_index(pd.to_datetime(underlying_prc_his_Tushare['trade_date'], format='%Y%m%d').dt.strftime('%Y-%m-%d'))
    underlying_prc_his.sort_index(inplace=True)
# 设置手续费和佣金
if underlying_symbol in index_list:
    handling_fee = 15
    commission = 15*3
elif underlying_symbol in etf_list:
    handling_fee = 1.3+0.3
    commission = 5    
total_fee = handling_fee + commission
```

```python
def get_tenth_of_month(start_date, end_date, date_select):
    start_date , end_date = pd.to_datetime(start_date), pd.to_datetime(end_date)
    date_list = []
    if start_date.day > date_select:  # 如果开始日期已经过了10th，调整为下月10th
        current_date = datetime.datetime(start_date.year, start_date.month, date_select) + relativedelta(months=1)
    else:
        current_date = datetime.datetime(start_date.year, start_date.month, date_select)  # 将开始日期调整为该月的10号
    while current_date <= end_date:  # 循环遍历每个月，直到当前日期大于等于结束日期
        date_list.append(current_date)
        current_date += relativedelta(months=1)
    return date_list

def get_adj_date(date_list):  # 将非交易日顺延
    adj_date = []
    trade_date_list = hq.get_trade_date(start_date=date_list[0], end_date=(date_list[-1] + relativedelta(months=1)), only_list=True)  # 获取区间内交易日，延长一月防止最后一个日期是非交易日
    for date in date_list:
        if date in trade_date_list:  # 如果列表内日期是交易日，不调整
            adj_date.append(date)
        else:  # 将trade_date_list里所有的日期与当前日期做差，找最小的正值，加在当前日期上
            for trade_day in trade_date_list:
                delta = trade_day - date
                if delta > datetime.timedelta(0):
                    date = date + delta
                    break
            adj_date.append(date)
    return adj_date

#  设置保证金参数
def calculate_margin(code, date, margin_adjustment_coefficient=0.15, minimum_guarantee_coefficient=0.5) :  # 单张卖权的保证金计算
    date=date.strftime('%Y-%m-%d')
    daily_info = opt_info.get_daily_info(code=code, date=date)
    contract_info = opt_info.get_contract_info(code=code)
    contract_unit = contract_info['contract_unit'].iloc[0]
    settle_price = daily_info['settle_price'].iloc[0]
    exercise_price = contract_info['exercise_price'].iloc[0]
    underlying_price = underlying_prc_his.loc[date, 'close']
    if contract_info['contract_type'].iloc[0] == 'CO':  # 看涨期权内在价值
        intrinsic_value = max((underlying_price - exercise_price) * contract_unit, 0)
    elif contract_info['contract_type'].iloc[0] == 'PO':  # 看跌期权内在价值
        intrinsic_value = max((exercise_price - underlying_price) * contract_unit, 0)

    margin = (settle_price * contract_unit
            + max(
            underlying_price * contract_unit * margin_adjustment_coefficient - intrinsic_value,
            minimum_guarantee_coefficient * underlying_price * contract_unit * margin_adjustment_coefficient
                )
            )
    return margin

open_date_list = get_adj_date(get_tenth_of_month(start_date, end_date, date_select))
close_date_list = get_adj_date(get_tenth_of_month(start_date, end_date, date_select))[1:]
#print(open_date_list)
#print(len(open_date_list))
#print(close_date_list)
#print(len(close_date_list))

def get_exercise_ym(date, relative_month):
    '''
    '''
    date = pd.to_datetime(date)
    return (date + relativedelta(months=relative_month)).strftime("%Y-%m")
```

```python
def find_target_opt(temp_info, underlying_pre_close, price_gear):
    exercise_price_gears = temp_info['exercise_price'].tolist()
    low = 0
    high = len(exercise_price_gears) - 1
    while low <= high:
        mid = (low + high) // 2
        mid_value = exercise_price_gears[mid]
        if mid_value < underlying_pre_close:
            low = mid + 1
        else:
            high = mid - 1
    if price_gear <0:
        target_gear = high+price_gear+1    
    elif price_gear >0:
        target_gear = low+price_gear-1
    return temp_info.iloc[target_gear,:]
```

```python
# 设置订单簿
orderbook = pd.DataFrame(columns=['datetime', 'code', 'trade_price', 'hold_price', 'trade_num', 'res_num', 'contract_unit', 'init_margin'])
```

```python
def open_order(open_date, trade_num, price_gear, contract_type):
    order = pd.DataFrame(index=[0], columns=['datetime', 'code', 'trade_price', 'hold_price', 'trade_num', 'res_num', 'contract_unit', 'init_margin'])
    underlying_pre_close = underlying_prc_his.loc[open_date.strftime('%Y-%m-%d'), 'pre_close']  # 寻找开仓日前一日收盘价
    #print(underlying_pre_close)
    exercise_ym = get_exercise_ym(open_date, time_gear)
    full_opt_info = opt_info.get_gear_info(underlying_symbol=underlying_symbol_for_opt, contract_type=contract_type, exercise_ym=exercise_ym, ascending=True, date=open_date.strftime('%Y-%m-%d'))
    code = find_target_opt(full_opt_info, underlying_pre_close, price_gear)['code']
    contract_unit = find_target_opt(full_opt_info, underlying_pre_close, price_gear)['contract_unit']
    daily_info = opt_info.get_daily_info(code=code, date=open_date.strftime('%Y-%m-%d'))
    
    order['datetime'].iloc[0] = open_date 
    order['code'].iloc[0] = code  
    
    order['hold_price'].iloc[0] = daily_info['close'].iloc[0]*contract_unit
    order['trade_num'].iloc[0] = trade_num
    order['res_num'].iloc[0] = order['trade_num'].iloc[0]
    order['contract_unit'] = contract_unit
    if trade_num >0:
        order['trade_price'].iloc[0] = daily_info['close'].iloc[0]*contract_unit + total_fee
    else:
        order['trade_price'].iloc[0] = daily_info['close'].iloc[0]*contract_unit - total_fee
    #print(order)
    return order

skip = 0
full_open_order = pd.DataFrame()
for open_date in open_date_list:
    if opt_info.check_trading(underlying_symbol = underlying_symbol_for_opt, date=open_date) == False:        
        skip +=1
        continue
    for i in range(0,len(trade_num)):        
        full_open_order = full_open_order.append(open_order(open_date, trade_num[i], price_gear[i], contract_type[i]), ignore_index=True)

#print(full_open_order)
```

```python
def close_order(open_order):
    order = pd.DataFrame(index=[0], columns=['datetime', 'code', 'trade_price', 'hold_price', 'trade_num', 'res_num', 'contract_unit', 'init_margin'])
    #print(open_order)
    close_date = close_date_list[open_date_list.index(open_order['datetime'])]
    #print(close_date)
    daily_info= opt_info.get_daily_info(code=open_order['code'], date=close_date.strftime('%Y-%m-%d'))
    #print(daily_info)
    order['datetime'] = close_date
    order['code'] = open_order['code']       
    order['hold_price'] = open_order['contract_unit'] * daily_info['close'].iloc[0]
    order['trade_num'] = -open_order['res_num']
    order['res_num'] = open_order['res_num'] + order['trade_num']
    order['init_margin'] = 0
    order['contract_unit'] = open_order['contract_unit']
    if order['trade_num'].item() > 0:
        order['trade_price'].iloc[0] = open_order['contract_unit'] * daily_info['close'].iloc[0] + total_fee
    else:
        order['trade_price'].iloc[0] = open_order['contract_unit'] * daily_info['close'].iloc[0] - total_fee
    #print(order)
    
    return order

full_close_order = pd.DataFrame()
for i in range(0,len(full_open_order)):
    
    if open_date_list.index(full_open_order['datetime'].iloc[i]) <= len(close_date_list)-1:
        full_close_order = full_close_order.append(close_order(full_open_order.iloc[i]), ignore_index=True)
    else:
        continue
    
    
#print(full_close_order)
```

```python
orderbook = orderbook.append(full_open_order, ignore_index=True)
orderbook = orderbook.append(full_close_order, ignore_index=True)
orderbook.sort_values('datetime', inplace=True)
orderbook.reset_index(drop=True, inplace=True)
# print(orderbook)
```

```python
# 设置期权业绩表
performance = pd.DataFrame(index=hq.get_trade_date(start_date=open_date_list[skip], end_date=end_date, only_list=True))
performance['opt_account_hold_value'] = 0
performance['opt_account_moving_accounts'] = 0
performance['opt_account_init_cash'] = init_cash
performance['opt_account_cash'] = 0
performance['opt_account_total_value'] = 0
performance['opt_account_margin'] = 0

for code in orderbook['code'].unique().tolist():
    code_rows = orderbook[orderbook['code'] == code]
    code_rows.set_index(pd.to_datetime(code_rows['datetime']), inplace=True)
    code_rows.drop('datetime', axis=1, inplace=True)
    #print(code_rows)
    start_date_period = code_rows.index[0]
    if len(code_rows.index) == 1:  # 仅开未平的情况
        end_date_period = end_date
    else:
        end_date_period = code_rows.index[1]
    
    trade_day_list = hq.get_trade_date(start_date=start_date_period, end_date=end_date_period, only_list=True)
    code_rows_filled = code_rows.reindex(trade_day_list)
    code_rows_filled['code'].fillna(code, inplace=True)
    code_rows_filled['trade_price'].fillna(0, inplace=True)
    code_rows_filled['trade_num'].fillna(0, inplace=True)
    code_rows_filled['res_num'].fillna(code_rows_filled['res_num'].iloc[0], inplace=True)
    #print(code_rows_filled)
    
    for index in code_rows_filled.index: # 通过数据库获取当天该合约的收盘价,计算持仓价值和保证金
        
        daily_info = opt_info.get_daily_info(code=code_rows_filled['code'].loc[index], date=index.strftime('%Y-%m-%d'))
        contract_unit = opt_info.get_contract_info(code=code_rows_filled['code'].loc[index])['contract_unit'].iloc[0]
        code_rows_filled['hold_price'].loc[index]= daily_info['close'].iloc[0]*contract_unit
        if code_rows_filled['res_num'].loc[index] < 0 :                        
            code_rows_filled['init_margin'].loc[index]= -code_rows_filled['res_num'].loc[index]*calculate_margin(code_rows_filled['code'].loc[index], index)
        else:
            code_rows_filled['init_margin'].loc[index]=0
        performance.loc[index, 'opt_account_margin'] += code_rows_filled['init_margin'].loc[index]
        performance.loc[index, 'opt_account_moving_accounts'] -= code_rows_filled.loc[index, 'trade_price']*code_rows_filled.loc[index, 'trade_num']
        performance.loc[index, 'opt_account_hold_value'] += code_rows_filled.loc[index, 'hold_price'] * code_rows_filled.loc[index, 'res_num']
    #print(code_rows_filled)
    #print(performance)

performance['opt_account_cash'].iloc[0] = performance['opt_account_init_cash'].iloc[0] + performance['opt_account_moving_accounts'].iloc[0]# 'cash'列第一行的值为'init_cash'列第一行的值
for i in range(1, len(performance)):  # 更新'cash'列其余行的值等于上一天cash的值加当天movingaccount的值
    performance['opt_account_cash'].iloc[i] = performance['opt_account_cash'].iloc[i-1] + performance['opt_account_moving_accounts'].iloc[i]
performance['opt_account_total_value'] = performance['opt_account_cash'] + performance['opt_account_hold_value']
performance['opt_account_net_value'] = performance['opt_account_total_value'] / performance['opt_account_init_cash'].iloc[0]
# print(performance)
       
```

```python
plt.plot(performance['opt_account_net_value'], label = 'opt_account_net_value')
plt.legend()
plt.show()
```
